```python
from llama_index.core import Document
from llama_index.core.node_parser import MarkdownNodeParser
from llama_index.core.node_parser import (
    SentenceSplitter,
    SemanticSplitterNodeParser,
)
from llama_index.embeddings.openai import OpenAIEmbedding
import os

os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
```

```python
with open("./data/Controller_Brochure.md", "r", encoding="utf-8") as f:
    markdown_content = f.read()

# Tạo một đối tượng Document
document = Document(text=markdown_content)

embedding = OpenAIEmbedding()

# 3. Lấy các node (đoạn) từ tài liệu
splitter = SemanticSplitterNodeParser(
    buffer_size=1,  # Số câu đệm giữa các chunk
    breakpoint_percentile_threshold=95,
    embed_model=embedding,
)

nodes = splitter.get_nodes_from_documents([document])

# 4. In ra các đoạn đã được chia
for i, node in enumerate(nodes):
    print(f"--- Semantic Chunk {i+1} ---")
    print(node.metadata)
    print(node.get_content())
    # print(f"Similarity to next: {node.metadata.get('next_chunk_similarity')}")
```
